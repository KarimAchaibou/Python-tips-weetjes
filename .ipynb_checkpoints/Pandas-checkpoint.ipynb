# Query methode
The ‘query()’ method provides an easy way to select data using string expressions. Let’s demonstrate the use of this method using the same data from above.

```python
df_Q = pd.read_csv("dow_jones.csv")
df_Q.info()
```

```python
#convert the date column into a datetime object
df_Q["Date"] = pd.to_datetime(df_Q["Date"])
df_Q.info()
```

```python
#let’s create month and week columns:
df_Q["month"] = df_Q["Date"].dt.month
df_Q["week"] = df_Q["Date"].dt.week
df_Q["year"] = df_Q["Date"].dt.year
df_Q
```

```python
#let’s use the ‘query()’ method to select data based on the week and month:
df_Q1 = df_Q.query("week<9")
df_Q1
```

```python
df_Q1 = df_Q1.query("month==2")
df_Q1
```

# The Apply Method

The next method we will discuss is the ‘apply()’ method. Suppose we wanted to take the square of the ‘Open’ column. To demonstrate this using ‘apply(), 

```python
# let’s first remove the comma from the string values in the ‘Open’ column 
# and convert the string values into a float:
df_Q["Open"] = df_Q["Open"].str.replace(",","")
df_Q["Open"] = df_Q["Open"].astype(float)
df_Q.info()
```

```python
df_A = df_Q
df_A.head()
```

```python
df_A["Open^2"] = df_A["Open"].apply(lambda x: x**2)
df_A.head()
```

A more practical use case is using the apply method with a custom function. Suppose we want to calculate returns from the ‘Open’ and ‘Close*’ values. Let’s remove the comma from the string values in the ‘Close*’ column and convert the string values into a float:

```python
df_A["Close*"] = df_A["Close*"].str.replace(",", "")
df_A["Close*"] = df_A["Close*"].astype(float)
df_A.head()
```

```python
# let’s define a function that we will use to calculate returns
def calculate_returns(df_in):
    return ((df_in[1]-df_in[0])/df_in[0])*100
```

```python
df_A["returns"] = df_A[["Open","Close*"]].apply(calculate_returns,axis=1)
df_A.head()
```

# Slicing
Slice notation being [start:stop:step]

.loc uses label based indexing to select both rows and columns. The labels being the values of the index or the 
columns. Slicing with .loc includes the last element. 

# Different choices for indexing

Object selection has had a number of user-requested additions in order to support more explicit location based indexing. Pandas now supports three types of multi-axis indexing.

    .loc is primarily label based, but may also be used with a boolean array. .loc will raise KeyError when the items are not found. Allowed inputs are:

            A single label, e.g. 5 or 'a' (Note that 5 is interpreted as a label of the index. This use is not an integer position along the index.).

            A list or array of labels ['a', 'b', 'c'].

            A slice object with labels 'a':'f' (Note that contrary to usual python slices, both the start and the stop are included, when present in the index! See Slicing with labels and Endpoints are inclusive.)

            A boolean array (any NA values will be treated as False).

            A callable function with one argument (the calling Series or DataFrame) and that returns valid output for indexing (one of the above).

    See more at Selection by Label.

    .iloc is primarily integer position based (from 0 to length-1 of the axis), but may also be used with a boolean array. .iloc will raise IndexError if a requested indexer is out-of-bounds, except slice indexers which allow out-of-bounds indexing. (this conforms with Python/NumPy slice semantics). Allowed inputs are:

            An integer e.g. 5.

            A list or array of integers [4, 3, 0].

            A slice object with ints 1:7.

            A boolean array (any NA values will be treated as False).

            A callable function with one argument (the calling Series or DataFrame) and that returns valid output for indexing (one of the above).

    See more at Selection by Position, Advanced Indexing and Advanced Hierarchical.

    .loc, .iloc, and also [] indexing can accept a callable as indexer. See more at Selection By Callable.
